**Simple Linear Regression - Model Assumptions:**



1.   Linear relationship:


       The regression model is linear in parameters

2.   Normality of Residuals Term :


        Confidence intervals may become too wide or narrow if the error components are not normally distributed. Confidence interval instability makes it challenging to estimate coefficients using the minimization of least squares.

3.   Multicollinearity:
        
        
        It is expected that the data are not multicollinear. When the features—or independent variables—are substantially connected, multicollinearity results.


4.   Auto-correlation:


        It is expected that the data should not have auto-correlation. When the residual errors are not unrelated to one another, autocorrelation happens.


5.   The number of observations must be greater than no of regressors:


6.   Observations of the error term are uncorrelated with each other:


          If there is a correlation inside the error term data, that we will examine. The observations are not picked at random if this occurs.

7.   All independent variables are uncorrelated with the error term:


      Your linear regression model should be checked to see whether any independent variables and the error term are correlated. If this occurs, you probably have a model that was incorrectly stated. You could have overlooked a crucial explanatory factor.

8.  The Expected value of error term is zero.
     
        Mathematically, E(ei) = 0 

9.  Homoscedasticity:

        Heteroskedasticity is a result of the error terms having non-constant variance. Non-constant variance often occurs when there are outliers or extremely high leverage values. It appears that these parameters are given an excessive amount of weight, which adversely affects the model's performance. The confidence interval for an out-of-sample prediction tends to be overly large or tight when this occurrence happens.



**Ordinary Least Squares (OLS)**

In statistics, ordinary least squares (OLS) is a type of linear least squares method for choosing the unknown parameters in a linear regression model  by the principle of least squares: minimizing the sum of the squares of the differences between the observed dependent variable  function of the independent variable.

**Error (Cost Function):**

The regression line must closely reflect the dataset. To put it differently, we want the line to be as near as feasible to the real data points. By reducing the vertical separation between the actual data point and fitted line, it can be accomplished. Each data point's vertical separation from the line is measured. The residual refers to this separation.
By identifying the line of best fit, we may reduce the residuals in a regression model. The vertical dotted lines connecting the actual data points to the line indicate the residuals.
A high positive residual would cancel out a large negative residual if we attempted to minimise the total of the residuals.We reduce the sum of the squares of the residuals as a result.

The total error present in the model which is the sum of the total errors of each individual data point known as Cost function.

Mathematically, the residual for a data point i would be given as di = yi - ŷi.
               where yi = be actual data points
                     ŷi = be predicted data points

Sum of the squares of the residuals is given as:

			D = Ʃ di**2       for all data points

By minimizing the model's error i.e., by minimizing D, we can estimate the model's parameters β0 and β1. We can locate the regression line as a result.

Ordinary Least Square Method is the name of the technique used to determine the model's parameters and, consequently, the regression line.


**Regression metrics for model performance:**

Evaluate model performance.

For regression problems, there are two ways to compute the model performance. They are RMSE (Root Mean Square Error) and R-Squared Value. These are explained below:-

**R2:**

R2 Score is metric to evaluate performance of a regression model. It is also called coefficient of determination. It gives us an idea of goodness of fit for the linear regression models. It indicates the percentage of variance that is explained by the model.

Mathematically,

        R2 Score = Explained Variation/Total Variation


*Conclusion: The higher the R2 Score value, the better the model fits the data. Usually, 
its value ranges from 0 to 1. So, we want its value to be as close to 1. Its value can become negative if our model is wrong.

**Root Mean Square Error:**

RMSE is the standard deviation of the residuals. So, RMSE gives us the standard deviation of the unexplained variance by the model. It can be calculated by taking square root of Mean Squared Error. RMSE is an absolute measure of fit. It gives us how spread the residuals are, given by the standard deviation of the residuals. The more concentrated the data is around the regression line, the lower the residuals and hence lower the standard deviation of residuals. It results in lower values of RMSE. So, lower values of RMSE indicate better fit of data.

**Error Analysis:**


A linear regression model might not accurately portray the data. The data and the model might not match well. As a result, we should define and evaluate residual plots to verify our model.

The residual, represented by the letter e, is the variation between the dependent variable's observed value (y) and the anticipated value I

The residual plot is the scatter-plot of these residuals.

A linear regression model could be acceptable for the data if the data points in a residual plot are randomly distributed across the horizontal axis and have a roughly zero residual mean. If not, a non-linear model would be more suitable.

The resulting "Residual errors" plot makes it abundantly evident that the train data plot pattern is not random. The test data plot pattern is in the same situation. Thus, it implies that a non-linear model would fit the data better.

````python
# Plotting the Error
d = [i for i in range(1,len(y_test)+1,1)]
plt.figure(figsize=(25,25))
plt.plot(d,y_test-predection_01,color='red',linestyle='-')
plt.xlabel('index')
plt.ylabel('Error')
plt.title('Error Value')
plt.show()
````

**Error Analysis:**


A linear regression model might not accurately portray the data. The data and the model might not match well. As a result, we should define and evaluate residual plots to verify our model.

The residual, represented by the letter e, is the variation between the dependent variable's observed value (y) and the anticipated value I

The residual plot is the scatter-plot of these residuals.

A linear regression model could be acceptable for the data if the data points in a residual plot are randomly distributed across the horizontal axis and have a roughly zero residual mean. If not, a non-linear model would be more suitable.

The resulting "Residual errors" plot makes it abundantly evident that the train data plot pattern is not random. The test data plot pattern is in the same situation. Thus, it implies that a non-linear model would fit the data better

````python
# Plotting for residual Error
plt.figure(figsize=(25,25))
plt.scatter(lr.predict(x_train), lr.predict(x_train) - y_train, c = 'blue', s = 40, label = 'Train data', alpha = 0.5)
plt.scatter(lr.predict(x_test), lr.predict(x_test) - y_test, c = 'orange', s = 40, label = 'Test data')
plt.hlines(y = 0, xmin = -100, xmax = 1000, linewidth = 3)
plt.legend(loc = 'upper right')
plt.title("Residual errors plot")
plt.show()
````

**Interpretation Residual Plot**

Analysis of residual plots
The residuals of a regression model that has successfully fit the data will show unpredictability (i.e., lack of any pattern). This is a result of regression modeling's homoscedasticity assumption. To verify the assumption, scatter plots between residuals and predictors are frequently utilised. Any pattern in the scatter-plot is a breach of this condition and indicates a model that doesn't match the data well.


The residual errors graphic demonstrates how the data is dispersed at random around line zero. There is no discernible pattern in the residuals on the plot. As a result, we may say that the data fit the linear regression model well.
